# 🏠 Lets Build a House: Understanding Python Classes and Objects

## 🧱 Introduction

In this notebook, we will learn how Python classes and objects work using a simple real-world metaphor: **houses in a neighborhood**.

### Key Idea:
- A **class** is a blueprint for a house
- An **object** is an actual house built from that blueprint
- Each house can have different features, but they follow the same design

## 🧱 The Class

```python
class House:
    def __init__(self, bedrooms, color):
        # self = the specific house being created
        self.bedrooms = bedrooms
        self.color = color

    def describe(self):
        print(f"This house has {self.bedrooms} bedrooms and is {self.color}.")

    def repaint(self, new_color):
        self.color = new_color
        print(f"The house is now {self.color}.")
```

## 🧱 The Class = Blueprint

Think of this `House` class as a **blueprint for building houses**.

It defines:
- What every house has (bedrooms, color)
- What every house can do (describe itself, change color)

But right now — no houses exist yet.

## 🏠 Create Objects

```python
house1 = House(2, "blue")
house2 = House(4, "red")
```

## 🏠 Objects = Real Houses

Now we have created two real houses:

- house1 → 2 bedrooms, blue
- house2 → 4 bedrooms, red

Even though they use the same blueprint, they are completely separate houses.

## 🚪 Use Methods

```python
house1.describe()
house2.describe()
```

## 🚪 Methods = Actions Houses Can Do

Each house can:
- describe itself
- change its color

These actions come from the blueprint (class), but are used by each individual house.

## 🎨 Modify One Object

```python
house1.repaint("green")
house1.describe()
```

## 🔁 Objects Are Independent

When we changed house1, house2 did NOT change.

This is because each object has its own separate data stored in memory.

## 🏠 Final Comparison

```python
house1.describe()
house2.describe()
```

## 🧠 Final Summary

### Class:
A blueprint that defines structure and behavior.

### Object
A real instance created from that blueprint.

### self
Refers to the specific object being worked on.

---

## Big Idea:
One class can create many independent objects, each with its own unique data.
