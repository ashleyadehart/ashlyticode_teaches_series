# 🏠 Lets Build a House: Understanding Python Classes and Objects

## Introduction

In this notebook, we will learn how Python classes and objects work using a simple real-world metaphor: **houses in a neighborhood**.

### Key Idea(s):
- A **class** is a blueprint for a house.
- An **object** is an actual house built from that blueprint.
- Each house can have different **attributes** (features), but they follow the same design.

## Designate Class

```python
# Designate class for the house
class House:
    def __init__(self, bedrooms, color):
        # self = the specific house being created
        self.bedrooms = bedrooms
        self.color = color

    def describe(self):
        print(f"This house has {self.bedrooms} bedrooms and is {self.color}.")

    def repaint(self, new_color):
        self.color = new_color
        print(f"The house is now {self.color}.")
```

## The Class = Blueprint

Think of this `House` class as a **blueprint for building houses**.

It defines:
- What every house has (bedrooms, color)
- What every house can do (describe itself, change color)

But right now — no houses exist yet.

## Create Objects

```python
# Create objects for the House class
house1 = House(2, "blue")
house2 = House(4, "red")
```

## Objects = Real Houses

Now we have created two real houses:

- house1 → 2 bedrooms, blue
- house2 → 4 bedrooms, red

Even though they use the same blueprint, they are completely separate houses.

## Specifying Attributes

```python
# Specify the attributes of the house objects
print(house1.bedrooms)  # Output: 2
print(house1.color)     # Output: "blue"
```

## Attributes = Features of the House
Attributes are pieces of data stored inside an object.

Think of attributes as:
- characteristics
- properties
- or features

of an object.

A real house has attributes such as:
- color
- number of bedrooms
- number of bathrooms
- garage size

In Python objects, attributes work the same way.

Each object stores its own attribute values.

## Modifiying Attributes

```python
# Change the color of house1
house1.color = "green"

# Print updated color
print(house1.color)

# house2 is unchanged
print(house2.color)
```

## Attributes vs Methods

Attributes store data about an object.

Examples:
- bedrooms
- color
- garage

---

Methods are actions an object can perform.

Examples:
- describe()
- repaint()
- open_door()

## Use Methods

```python
class House:
    def __init__(self, bedrooms, color):
        self.bedrooms = bedrooms
        self.color = color

    def describe(self):
        print(f"This house has {self.bedrooms} bedrooms and is {self.color}.")

    def repaint(self, new_color):
        self.color = new_color
        print(f"The house is now {self.color}.")
```

## Methods = Actions Houses Can Do

Each house can:
- describe itself
- change its color

These actions come from the blueprint (class), but are used by each individual house.

## Modify Objects

```python
# Modify the House class to include a method that calculates the total number of rooms (bedrooms + 1 for the living room)
class House:
    def __init__(self, bedrooms, color):
        self.bedrooms = bedrooms
        self.color = color

    def describe(self):
        print(f"This house has {self.bedrooms} bedrooms and is {self.color}.")

    def repaint(self, new_color):
        self.color = new_color
        print(f"The house is now {self.color}.")

    def total_rooms(self):
        return self.bedrooms + 1  # +1 assumes 1 living room

house1 = House(2, "blue")
house2 = House(4, "red")
house3 = House(3, "yellow")

house1.describe()
house2.describe()
house3.describe()
```

## Lesson Summary
**Class**: A blueprint that defines structure and behavior.

**Object**: A real instance created from that blueprint.

**Attributes**: Pieces of data stored inside an object.

**Self**: Refers to the specific object being worked on.
